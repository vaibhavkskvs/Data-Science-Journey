# If statement

if <condition>:
    <code>
elif <condition>:
    <code>
...
else:
    <code>

```python
age = 18

if age>=18:
    print("You're an adult!")
else:
    print("You're a kid")
```

```python
age = 16

if age>=18:
    print("You're an adult!")
elif age>=13:
    print("You're a teenager")
else:
    print("You're a kid")
```
